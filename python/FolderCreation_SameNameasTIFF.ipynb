```python
import os
from PIL import Image

def create_folders_for_tiff(input_folder):
    # Walk through the directory and its subdirectories
    for root, dirs, files in os.walk(input_folder):
        for file in files:
            if file.lower().endswith('.tiff'):
                # Full path to the TIFF file
                tiff_path = os.path.join(root, file)
                
                # Create a new folder with the same name as the TIFF file (without extension)
                folder_name = os.path.splitext(file)[0]
                new_folder_path = os.path.join(root, folder_name)
                os.makedirs(new_folder_path, exist_ok=True)
                
                # Move the TIFF file into the new folder
                new_tiff_path = os.path.join(new_folder_path, file)
                os.rename(tiff_path, new_tiff_path)
                
                # Optionally, you can perform additional processing here (e.g., image manipulation)
                
                print(f"Processed: {file}")

if __name__ == "__main__":
    input_folder = r'X:\NOG_drive\20240128'
    create_folders_for_tiff(input_folder)
```
