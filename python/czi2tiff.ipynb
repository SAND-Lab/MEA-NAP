```python
import czifile
import tifffile
import numpy as np
import napari
```

```python
def check_czi_datatype(file_path):
    """Check the data type of a CZI file"""
    with czifile.CziFile(file_path) as czi:
        data_type = czi.dtype
    return data_type

def normalize(image):
    """Normalize an image array to 16-bit range"""
    image = image.astype(np.float32)
    image -= image.min()
    image /= image.max()
    image *= 2**16 - 1
    return image.astype(np.uint16)
    
def load_czi_file(file_path):
    """Load a CZI file and return the image data and metadata"""
    with czifile.CziFile(file_path) as czi:
        metadata = czi.metadata()
        image = czi.asarray()
    return image, metadata

def save_tiff_file(image, file_path):
    """Save a NumPy array as a TIFF file"""
    tifffile.imsave(file_path, image)

def view_czi(file_path): 
    viewer = napari.Viewer()
    viewer.open(file_path)

def get_czi_dimensions(file_path):
    """Get the dimensions of a CZI file"""
    with czifile.CziFile(file_path) as czi:
        dims = czi.shape
    return dims
```
